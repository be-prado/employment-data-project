## Compare income and interest rates for borrowing money

No one can say for sure, but what do you think will happen to
interest rates for borrowing money during the next 12
months--will they go up, stay the same, or go down?

- 1:	Go up
- 3:	Stay the same
- 5:	Go down
- 8:	DK
- 9:	NA

Treat '8','9' AS NA

INCOME	
TOTAL HOUSEHOLD INCOME - CURRENT DOLLARS	
Now, thinking about your total income from all sources
(including your job), how much did you receive in the
previous year?


```python
df1 = survey_df[['RATEX', 'INCOME']]
df1.head()
```

```python
df1['RATEX'].unique()
```

```python
def rate_groups(series):
    rate={
        1:'Go up',
        3:'Stay the same',
        5:'Go down',
        8:'DK',
        9:'NA'
        }
    return rate.get(series)

df1['rate'] = df1['RATEX'].apply(rate_groups)

df1['rate'].value_counts(sort=False)
```

```python
import numpy as np
df1['RATEX1']= df1['RATEX'].astype({'RATEX':'category'}).replace([8,9], np.nan)
```

```python
df1.describe(include='all')
```

```python
df1_full=df1.dropna()
df1_full.describe(include='all')
```

```python

import plotly.express as px
fig = px.box(df1_full, x="rate", y="INCOME",width=600, height=600,
             category_orders={"rate":["Go up","Stay the same","Go down"]})
fig.update_layout(title_text='Income and Interest Rates for Borrowing Money')
fig.show()
```

people having a positive attitude about interest rate tend to be income median 32.5k, while others that higher income tend to believe up/stay the same.

1: up 3: stay 5: down

## political affiliation by education level

Generally speaking, do you usually think of
yourself as a Republican, a Democrat, an Independent, or
what?
* 1:	Republican
* 2:	Democrat
* 3:	Independent, No Preference
* 8:	DK
* 9:	NA


EDUC	
Education of Respondent
* 1:	Grade 0-8 no hs diploma
- 2:	Grade 9-12 no hs diploma
- 3:	Grade 0-12 w/ hs diploma
- 4:	Grade 13-17 no col degree
- 5:	Grade 13-16 w/ col degree
- 6:	Grade 17 W/ col degree



```python
df2=survey_df[['POLAFF','EDUC']]
df2.describe(include='all')
```

```python
def politic_groups(series):
    rate={
        1:'Republican',
        2:'Democrat',
        3:'Independent, No Preference',
        8:'DK',
        9:'NA'
        }
    return rate.get(series)

df2['political'] = df2['POLAFF'].apply(politic_groups)

df2['political'].value_counts(sort=False)

```

```python
def educ_groups(series):
    rate={
        1:'Grade 0-8 no hs diploma',
        2:'Grade 9-12 no hs diploma',
        3:'Grade 0-12 w/ hs diploma',
        4:'Grade 13-17 no col degree',
        5:'Grade 13-16 w/ col degree',
        6:'Grade 17 W/ col degree'
        }
    return rate.get(series)

df2['education'] = df2['EDUC'].apply(educ_groups)

df2['education'].value_counts()
```

```python
df2_nan= df2.replace([8,9,None], np.nan)
# df2=pd.concat([df2[['EDUC']], dfPOLAFF], axis=1)
df2_full=df2_nan.dropna()
df2_full.head()
```

```python
df2_full.describe(include='all') #62445
```

```python
#df2_counts=df2_full.value_counts().unstack()  
#fig = df2_counts.T.plot(kind='bar',stacked='true')
```

```python
df2_value_counts=df2_full.value_counts().to_frame().reset_index()
df2_value_counts.columns = ['EDUC','POLAFF','education level', 'political affiliation','count'] # change column names
df2_value_counts
```

```python
fig = px.bar(df2_value_counts, x='political affiliation', y="count", color='EDUC', text='education level',
             category_orders={"political affiliation":['Republican','Democrat','Independent, No Preference'],
                              "education level":['Grade 0-8 no hs diploma','Grade 9-12 no hs diploma','Grade 0-12 w/ hs diploma','Grade 13-17 no col degree','Grade 13-16 w/ col degree','Grade 17 W/ col degree']})
fig.update_layout(title_text='Political Affiliation by Education Level') #, xaxis_tickangle=-45
fig.show()
```

From the bar chart above, we can see that most of the people have no preference regarding political affiliation. Most of the people in Grade 17 are Democrats, comparing to Republican or Independent. Most of the people in Grade 13-17 without college degree, or Grade 13-16 with college degree are Independents.
Most of the people in Grade 0-12 with high school diploma are Independents.
